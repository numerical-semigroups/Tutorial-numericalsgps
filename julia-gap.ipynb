# `julia` and `numericalsgps`

We will need the packages `GAP` and `Plot`

```julia
using GAP
```

```julia
G = GAP.Globals
```

```julia
to_gap = GAP.julia_to_gap
from_gap = GAP.gap_to_julia
gap_eval = GAP.EvalString
```

```julia
G.LoadPackage(to_gap("num"))
```

```julia
s = G.NumericalSemigroup(5,7,9)
```

```julia
G.FrobeniusNumber(s)
```

```julia
pr=G.MinimalPresentation(s)
```

```julia
pr[1]
```

```julia
length(pr)
```

```julia
nci(n) = length(G.CompleteIntersectionNumericalSemigroupsWithFrobeniusNumber(n))
```

```julia
nci(11)
```

```julia
using Plots
```

```julia
plot(nci, 1:2:51)
```

```julia
bar(nci,1:2:51,label="")
```

```julia
nf(n,s) = length(G.Factorizations(n,s))
```

```julia
s = G.NumericalSemigroup(11,13,19,21)
```

```julia
nf(100,s)
```

```julia
fst = G.FirstElementsOfNumericalSemigroup(100,s);
```

```julia
nfst = [nf(n,s) for n in fst];
```

```julia
bar([x for x in fst], nfst,label="")
```
