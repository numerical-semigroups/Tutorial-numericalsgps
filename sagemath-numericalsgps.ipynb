# Using `numericalsgps` in `SageMath`

We show some examples of using `sagemath`with `numericalsgps`.
After installing `numericalsgps`, we load it; beware that you need double quotes.

```python
gap.LoadPackage('"num"')
```

Let us see how many factorizations have the first 100 elements in $\langle 10,13,19,21 \rangle$, and let us plot the result.

```python
s = gap.NumericalSemigroup(10,13,19,21)
```

```python
l = gap.FirstElementsOfNumericalSemigroup(100,s)
```

```python
nf = [len(gap.Factorizations(n,s)) for n in l]
```

```python
list_plot(nf)
```

This semigroup has Frobenius number 37.

```python
f = gap.FrobeniusNumber(s)
```

```python
f
```

We can choose those elements with a unique expression.

```python
[n for n in l if len(gap.Factorizations(n,s))==1]
```

We can also write some functions using `gap` functionalities.

```python
def eliahou(n,s):
    gens = gap.MinimalGenerators(s)
    v = [g for g in gens if gap.BelongsToNumericalSemigroup(n-g,s)]
    e = [(a,b) for a in gens for b in gens if a<b and gap.BelongsToNumericalSemigroup(n-a-b,s)]
    g = graphs.EmptyGraph()
    g.add_vertices(v)
    g.add_edges(e)
    return g
```

```python
eliahou(300,s)
```

```python
gap.BettiElements(s)
```

```python
g=eliahou(40,s)
```

```python
g
```
